### Single Event

```python
from gtts import gTTS

# Your single event commentary text
text = "Dhoni finishes off in style! A magnificent strike into the crowd. India lift the World Cup after 28 years!"

# Initialize the gTTS object
tts = gTTS(text=text, lang='en')   #'en' stands for English language

# Save the generated speech to an .mp3 file
tts.save("commentary.mp3")
```

```python
from IPython.display import Audio

# Play the generated mp3 file
Audio("commentary.mp3")
```

### Multiple Events

```python
from gtts import gTTS
from IPython.display import Audio, display
import os

# List of 10 famous cricket commentary lines
commentary_lines = [
    "Dhoni finishes off in style! A magnificent strike into the crowd. India lift the World Cup after 28 years!",
    "What a delivery! Clean bowled — the stumps are shattered!",
    "Rohit Sharma flicks it gracefully over midwicket — that’s four more!",
    "It’s up in the air… and taken! A brilliant diving catch in the deep.",
    "Kohli drives through the covers — classic timing, races to the boundary.",
    "Pandya goes big! That’s a monstrous six over long-on.",
    "He’s run out! A direct hit from the boundary ends the partnership.",
    "Jadeja with a rocket arm and it’s a sharp run-out. That’s top fielding.",
    "Gill dances down the track and sends it into the second tier!",
    "India have won the match! A thrilling finish in the final over."
]

# Create directory to store MP3s
os.makedirs("commentary_audio", exist_ok=True)

# Generate and save audio for each commentary line
for i, line in enumerate(commentary_lines, start=1):
    tts = gTTS(text=line, lang='en')
    filename = f"commentary_{i}.mp3"
    tts.save(filename)
    print(f"✔️ Saved: {filename}")

    # Optional: Play audio in notebook
    display(Audio(filename))
```
